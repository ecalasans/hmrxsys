```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv('hmrxapp.csv')
df.head()
```

```python
df_eric = df[df['avaliador'] == 1]
```

```python
df_eric = df_eric.reset_index()
```

```python
df_eric.drop(columns=['index'])
```

```python
nome_arquivo = df_eric['arquivo'].reset_index()
```

```python
nome_arquivo = nome_arquivo.drop(columns=['index'])
```

```python
nome_arquivo
```

```python
histogramas = df_eric['histograma']
bins = [str(i) for i in range(256)]
histogramas.size
```

```python
h_series = []
for h in range(histogramas.size):
    histograma_pool = json.loads(histogramas.iloc[h])
    lista = [np.round(value, 2) for key, value in histograma_pool.items()]
    arr_hist = np.array(lista)
    h_series.append(arr_hist)
```

```python
h_series = np.array(h_series)
h_series.shape
```

```python
X = pd.DataFrame(h_series, columns=bins)
```

```python
X.head()
```

```python
Y = df_eric[['gamma_l', 'gamma_h', 'c', 'd0']]
Y.head()
```

```python
Y = Y.reset_index()
Y.drop(columns=['index'])
```

```python
analise_eric = pd.concat([X, Y, nome_arquivo], axis=1)
```

```python
analise_eric.shape
```

```python
analise_eric.columns
```

```python
analise_eric.drop(columns=['index'])
```

```python
analise_eric.to_csv('analise_eric.csv')
```
